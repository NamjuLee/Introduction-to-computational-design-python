```python
class Node:
    def __init__(self, idNumber):
        self.idNumber = idNumber
        self.connections = []
    def addNode(self, n0):
        self.connections.append(n0);
        n0.connections.append(self);
    def toJson(self):
        outDict = {}
        outDict['id'] = self.idNumber;
        outDict['connectivity'] = [];
        for i in self.connections:
            outDict['connectivity'].append(i.idNumber)
        return outDict
    def print(self):
        print("id: ", self.idNumber, ', numOfCon:', len(self.connections) )
```

### Graph class

```python
import random as rnd

class Graph:
    def __init__(self, num):
        self.nodes = [];
        for i in range(num):
            n = Node(i)
            self.nodes.append(n)
    def initCon(self, num):
        for i in range(num):
            index0 = rnd.randrange(0, len(self.nodes))
            index1 = rnd.randrange(0, len(self.nodes))
            if index0 != index1:
                self.nodes[index0].addNode(self.nodes[index1])
#                 print(index0, index1 )
    def toJson(self):
        outList = []
        for i in range(len(self.nodes)):
            outList.append(self.nodes[i].toJson())
        return outList;
```

```python
g = Graph(100)
g.initCon(1000)
theJson = g.toJson()
```
