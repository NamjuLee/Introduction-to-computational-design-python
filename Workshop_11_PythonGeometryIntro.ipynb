```python
class Line:
    array = [];
      
    def __init__(self, p0, p1):
        self.p0 = p0
        self.p1 = p1
        Line.array.append(self);
        
    def print(self):
        print("line by", 'p0:',self.p0,', p1:' , self.p1 )
```

```python
class Point3:
    array = [];
      
    def __init__(self, x, y, z = 0):
        self.x = x
        self.y = y
        self.z = z
        Point3.array.append(self);
        
    def __str__(self):
        return "x:" + str(self.x) + ", y:" + str(self.y) + ", z:" + str(self.z) 
```

```python
import math
def Linspace(start = -1, stop = 1, num = 100 ):
    interval = ( stop - start) / (num - 1);
    outArray = [];
    for i in range(num):
        outArray.append(start + (interval * i));
    return outArray;

def GenerateXYDataForTest():
    return GenerateXYData(- 100, 100, 30, [0.00006, -0.003, 0.1, 10]);

def GenerateXYData(theMin, theMax, divi, coeffs):
    data = [];
    interval = (theMax - theMin) / divi;
    for i in range(theMin, theMax):
        x = i * interval
        data.append([ x, coeffs[0] * x * x * x + coeffs[1] * x * x + coeffs[2] * x + coeffs[3]]);
    return data;

def GetGaussian(x, mu, sigma):
    return math.exp(-math.pow(x - mu, 2.0) / (2 * math.pow(sigma, 2.0)));

def GetGaussianEx():
    linspace = Linspace(-5, 5, 100);
    inData = [ [0, 1], [0, 2], [1, 0.5], [1, 0.75]];

    out = [];
    for d in inData:
        array = [];
        for x in linspace:
            y = GetGaussian(x, d[0], d[1]);
            array.append([x, y]);
        out.append(array);
    return out;
```

```python
def GenerateDataLinearFun(N=1000, beta0=2.2, beta1=3.0, sigma=10.0, Xmax=1.0):
    epsilon=np.random.normal(0,sigma,N)  # Random normally distributed points 
    X = np.linspace(0,Xmax, N)           
    Y = beta0 + beta1 * X + epsilon 
    return X, Y 


def GenerateDataNotLinearFun(N=1000, beta0=2.2, beta1=3.0, sigma=1.0, Xmax=1.0, alpha=1):
    epsilon=np.random.normal(0,sigma,N)  # Random normally distributed points 
    X = np.linspace(0,Xmax, N)           
    Y = beta0 + beta1 * X + alpha* np.sin(6*X)+ epsilon 
    #Y =  beta0 + beta1 * X +4*beta1*(X>0.5) + epsilon 
    #Y =  beta0 + beta1 * X + alpha * X*X*X + epsilon
    return X, Y 
```
